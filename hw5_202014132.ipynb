```python
chipo['order_id'] =chipo['order_id'].astype('str')
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
chipo.head()
```

```python
chipo.info()
```

```python
chipo.groupby('order_id')['item_price'].sum()
```

```python
chipo.groupby('order_id')['item_price'].sum().mean()
```

```python
chipo_odreid_group = chipo.groupby('order_id').sum()
results = chipo_odreid_group[chipo_odreid_group.item_price >=10]
print(results[:10])
print(results.index.values)
```

```python
chipo_one_item = chipo[chipo.quantity ==1]
price_per_item = chipo_one_item.groupby('item_name').min()
price_per_item.sort_values(by='item_price',ascending=False)[:10]
```

```python
chipo.groupby('order_id').sum().sort_values(by='item_price',ascending=False)
```
